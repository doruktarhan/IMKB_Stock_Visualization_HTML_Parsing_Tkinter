```python
# get the stock prices full names and symbols
from bs4 import BeautifulSoup
import requests
import pandas as pd

url = "https://en.wikipedia.org/wiki/List_of_companies_listed_on_the_Istanbul_Stock_Exchange"

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Find the first table on the page
stocks = soup.find_all('table',{'class':'wikitable sortable'})


#we have found the table so we have to get the elements of the table
companies_dict = {}
rows = stocks[0].find_all('tr')
for row in rows:
    columns = row.find_all('td')
    data = [col.text.strip() for col in columns]
    if len(data) > 1:
        companies_dict[data[0]] = data[1]

companies_dict
```

```python
# some of the stocks have multiple symbols so we have to fix them
for key,value in companies_dict.items():
    if ',' in value:
        companies_dict[key] = value.split(',', 1)[0]
    
companies_dict
```

```python
import json
with open('IMKB_Stocks.json','w',encoding='utf8') as f:
    json.dump(companies_dict,f,ensure_ascii=False)
```
